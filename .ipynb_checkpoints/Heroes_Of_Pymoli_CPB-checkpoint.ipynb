# Heroes of PyMoli

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data_HOP.csv"

# Read Purchasing File and store into Pandas data frame
game_data = pd.read_csv(file_to_load)
```

```python
#Visualize Data Set
game_data.head()
```

### Total Player Count

```python
#total player count
total_players = len(game_data["SN"].unique())
total_players_df = pd.DataFrame({"Total Players": [total_players]})
total_players_df
```

### Purchase Analysis

```python
#purchasing analysis
unique_items = len(game_data["Item ID"].unique())
avg_purchase_price = round((game_data["Price"].mean()), 2)
num_of_purchases = len(game_data["Purchase ID"].unique())
total_revenue = round((game_data["Price"].sum()), 2)
#create new DataFrame to store values
purchasing_analysis_df = pd.DataFrame({"Number of Unique Items": [unique_items],
                                       "Average Purchase Price": ["$" +str(avg_purchase_price)],
                                       "Total Number of Purchases": [num_of_purchases],
                                       "Total Revenue": ["$" +str(total_revenue)]})
purchasing_analysis_df
```

### Gender Demographics

```python
#Solution for Gender Demo
gender_df = game_data[["SN", "Gender"]]

#male demos
male_SNlist = gender_df.loc[gender_df["Gender"] == "Male", :]
male_count = len(male_SNlist["SN"].unique())

#female demos
female_SNlist = gender_df.loc[gender_df["Gender"] == "Female", :]
female_count = len(female_SNlist["SN"].unique())

#other demos
other_SNlist = gender_df.loc[gender_df["Gender"] == "Other / Non-Disclosed", :]
other_count = len(other_SNlist["SN"].unique())

#calculate percentages
male_pct = round((male_count/total_players*100), 2)
female_pct = round((female_count/total_players*100), 2)
other_pct = round((other_count/total_players*100), 2)

#create new DF to display gender demo values
gender_demo_df = pd.DataFrame({"Gender":["Male", "Female", "Other/Non-Disclosed"],
                               "Player Count":[male_count, female_count, other_count],
                               "Percentage of Playerbase":[str(male_pct) + "%", str(female_pct) + "%", str(other_pct) + "%"]})
gender_demo_df
```

### Purchase Analysis by Gender

```python
#create gender purchase analysis DF from total DF
gender_purchase_df = game_data[["Purchase ID","SN", "Gender", "Price"]]

#male purchase analysis
m_SN_purchase_list = gender_purchase_df.loc[gender_df["Gender"] == "Male", :]
male_pcount = len(m_SN_purchase_list["Purchase ID"].unique())
male_ptotal = round((m_SN_purchase_list["Price"].sum()), 2)
male_pavg = round((m_SN_purchase_list["Price"].mean()), 2)
male_ptotal_per_person = round((male_ptotal / male_count), 2)

#female purchase analysis
f_SN_purchase_list = gender_purchase_df.loc[gender_df["Gender"] == "Female", :]
female_pcount = len(f_SN_purchase_list["Purchase ID"].unique())
female_ptotal = round((f_SN_purchase_list["Price"].sum()), 2)
female_pavg = round((f_SN_purchase_list["Price"].mean()), 2)
female_ptotal_per_person = round((female_ptotal / female_count), 2)

#other purchase analysis
o_SN_purchase_list = gender_purchase_df.loc[gender_df["Gender"] == "Other / Non-Disclosed", :]
other_pcount = len(o_SN_purchase_list["Purchase ID"].unique())
other_ptotal = round((o_SN_purchase_list["Price"].sum()), 2)
other_pavg = round((o_SN_purchase_list["Price"].mean()),2)
other_ptotal_per_person = round((other_ptotal / other_count), 2)

#create new DF to store and display above values
gender_purchase_analysis_df = pd.DataFrame({
                                "Gender":["Male", "Female", "Other/Non-Disclosed"],
                                "Purchase Count":[male_pcount, female_pcount, other_pcount],
                                "Average Purchase Price":["$" + str(male_pavg), "$" + str(female_pavg), "$" + str(other_pavg)],
                                "Total Purchase Value":["$" + str(male_ptotal), "$" + str(female_ptotal), "$" + str(other_ptotal)],
                                "Average Purchase Per Player":["$" + str(male_ptotal_per_person), "$" + str(female_ptotal_per_person), "$" + str(other_ptotal_per_person)]})
gender_purchase_analysis_df
```

### Top Spenders

```python
#finding top spenders
top_spenders_df = game_data[["SN", "Price"]]

#get purchase count
ts_pcount_df = pd.DataFrame(top_spenders_df["SN"].value_counts())
ts_pcount_df.reset_index(inplace=True)
ts_pcount_df.columns = ["SN", "Purchase Count"]

#get average purchase value for each SN
ts_pavg_df = pd.DataFrame(top_spenders_df.groupby("SN")["Price"].mean())
ts_pavg_df.reset_index(inplace=True)
ts_pavg_df.columns = ["SN","Average Purchase Value"]

#get purchase total for each SN
ts_ptotal_df = pd.DataFrame(top_spenders_df.groupby("SN")["Price"].sum())
ts_ptotal_df.reset_index(inplace=True)
ts_ptotal_df.columns = ["SN","Total Purchase Value"]

#Merge above values into final DF
merged_spenders_df = pd.merge(ts_pcount_df, ts_pavg_df, on="SN")
merged_spenders2_df = pd.merge(merged_spenders_df, ts_ptotal_df, on="SN")
merged_spenders3_df = merged_spenders2_df.set_index("SN")

#Solution for Top Spenders
merged_spendersF_df = merged_spenders3_df.sort_values("Total Purchase Value",ascending=False).head(5)
#Formatted columns to correct currency formatting
merged_spendersF_df["Average Purchase Value"] = merged_spendersF_df["Average Purchase Value"].map("${0:,.2f}".format)
merged_spendersF_df["Total Purchase Value"] = merged_spendersF_df["Total Purchase Value"].map("${0:,.2f}".format)
merged_spendersF_df
```

### Most Popular Items

```python
#Find Most Popular Items via new DF
items_pop = game_data[["Item ID", "Item Name", "Price"]]

#Find count of each Item by Name
items_df = pd.DataFrame(items_pop["Item Name"].value_counts())
items_df.reset_index(inplace=True)
items_df.columns = ["Item Name", "Purchase Count"]

#isolate Total Purchase Value by Name for merge
items_tprice_df = pd.DataFrame(items_pop.groupby("Item Name")["Price"].sum())
items_tprice_df.reset_index(inplace=True)
items_tprice_df.columns = ["Item Name","Total Purchase Value"]

#isolate Price by Item Name for merge
items_price_df = pd.DataFrame(items_pop.groupby("Item Name")["Price"].mean())
items_price_df.reset_index(inplace=True)

#isolate Item ID by Item Name for merge
items_id_df = pd.DataFrame(items_pop.groupby("Item Name")["Item ID"].mean())
items_id_df.reset_index(inplace=True)

#Merged the above DF's
merged_items_df = pd.merge(items_df, items_price_df, on="Item Name")
merged_items2_df = pd.merge(merged_items_df, items_tprice_df, on="Item Name")
merged_items3_df = pd.merge(merged_items2_df, items_id_df, on="Item Name")
merged_items4_df = merged_items3_df.set_index("Item ID")

#Solution for Most Popular Items
merged_items_pop_df = merged_items4_df.sort_values("Purchase Count",ascending=False).head(5)
#Formatted columns to correct currency formatting
merged_items_pop_df["Total Purchase Value"] = merged_items_pop_df["Total Purchase Value"].map("${0:,.2f}".format)
merged_items_pop_df["Price"] = merged_items_pop_df["Price"].map("${0:,.2f}".format)
merged_items_pop_df
```

### Most Profitable Items

```python
#Solution for Most Profitable Items
merged_items_profit_df = merged_items4_df.sort_values("Total Purchase Value",ascending=False).head(5)
#Formatted columns to correct currency formatting
merged_items_profit_df["Total Purchase Value"] = merged_items_profit_df["Total Purchase Value"].map("${0:,.2f}".format)
merged_items_profit_df["Price"] = merged_items_profit_df["Price"].map("${0:,.2f}".format)
merged_items_profit_df
```
